```python
import matplotlib.pyplot as plt

# 1. Create our data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
ice_cream_sales = [10, 15, 45, 55, 80]
hot_chocolate_sales = [85, 70, 30, 15, 5]

# 2. Setup the figure and axes
fig, ax = plt.subplots(figsize=(8, 5)) # Width and height in inches

# 3. Plot multiple lines on the SAME axes
ax.plot(months, ice_cream_sales, label='Ice Cream', color='skyblue', marker='o', linewidth=2)
ax.plot(months, hot_chocolate_sales, label='Hot Chocolate', color='chocolate', linestyle='--', marker='s')

# 4. Customize the labels and titles using the "ax.set_" prefix
ax.set_title("Monthly Dessert Sales", fontsize=16, fontweight='bold')
ax.set_xlabel("Months", fontsize=12)
ax.set_ylabel("Units Sold", fontsize=12)

# 5. Add a legend (it reads the 'label' from step 3) and a background grid
ax.legend()
ax.grid(True, linestyle=':', alpha=0.6)

# 6. Show the masterpiece!
fig.savefig('monthly-dessert-sales', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a grid of 1 row and 2 columns
fig, ax = plt.subplots(1, 2, figsize=(10, 4), layout='constrained')

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Control the FIRST chart (Left side)
ax[0].plot(x, y, color='blue')
ax[0].set_title('Left Chart (Line)')

# Control the SECOND chart (Right side)
ax[1].bar(x, y, color='orange')
ax[1].set_title('Right Chart (Bar)')

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [10, 15, 35, 50] # Millions
staff = [50, 52, 80, 110]   # Raw count

# Plot revenue on the left axis
ax1.plot(months, revenue, color='g', label='Revenue')
ax1.set_ylabel('Revenue ($ Millions)', color='g')
ax1.tick_params(axis='y', labelcolor='g')

# Create a twin axis sharing the x-axis
ax2 = ax1.twinx()

# Plot staff on the right axis
ax2.plot(months, staff, color='b', linestyle='--', label='Staff')
ax2.set_ylabel('Staff Count', color='b')
ax2.tick_params(axis='y', labelcolor='b')

plt.show()
```
